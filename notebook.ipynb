```python
df = df.replace('dischromic _patches', 'dischromic_patches')
df = df.replace('spotting_ urination','spotting_urination')
df = df.replace('foul_smell_of urine','foul_smell_ofurine')
```

```python
df['Disease'].unique()
```

```python
# List of all diseases to include in the filter
diseases_to_filter = [
    'Fungal infection', 'Allergy', 'GERD', 'Chronic cholestasis',
    'Drug Reaction', 'Peptic ulcer diseae', 'AIDS', 'Diabetes',
    'Gastroenteritis', 'Bronchial Asthma', 'Hypertension', 'Migraine',
    'Cervical spondylosis', 'Paralysis (brain hemorrhage)', 'Jaundice',
    'Malaria', 'Chicken pox', 'Dengue', 'Typhoid', 'hepatitis A',
    'Hepatitis B', 'Hepatitis C', 'Hepatitis D', 'Hepatitis E',
    'Alcoholic hepatitis', 'Tuberculosis', 'Common Cold', 'Pneumonia',
    'Dimorphic hemmorhoids(piles)', 'Heart attack', 'Varicose veins',
    'Hypothyroidism', 'Hyperthyroidism', 'Hypoglycemia',
    'Osteoarthristis', 'Arthritis',
    '(vertigo) Paroymsal  Positional Vertigo', 'Acne',
    'Urinary tract infection', 'Psoriasis', 'Impetigo'
]

# Use the isin() method to create the condition
condition = df['Disease'].isin(diseases_to_filter)

# Apply the filter to the DataFrame
filtered_df = df[condition]

# Display the first few rows of the filtered DataFrame
filtered_df.head()
```

```python
# condition = (df['Disease'] == 'Fungal infection') | (df['Disease'] == 'Impetigo') | (df['Disease'] == 'Acne')
# filtered_df = df[condition]
# filtered_df.head()
```

```python
X = filtered_df.drop('Disease', axis=1)
y = filtered_df['Disease']
X.head()
```

```python
def merge_boolean_columns(data):
    merged_data = data.copy()
    unique_columns = set(data.columns)

    for column in unique_columns:
        columns_to_merge = data.columns[data.columns == column]
        if len(columns_to_merge) > 1:
            merged_column = data[columns_to_merge].any(axis=1)
            merged_data.drop(columns_to_merge, axis=1, inplace=True)
            merged_data[column] = merged_column
    return merged_data


for col in X.columns:
    encX = pd.get_dummies(X[col])
    X = pd.concat([X, encX], axis=1)
    X.drop(col, axis=1, inplace=True)
    
X_merged = merge_boolean_columns(X)

print(X.shape)
print(X_merged.shape)
```

```python
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder()
y_enc = label_encoder.fit_transform(y)

print("Class Mapping:")
for i, label in enumerate(label_encoder.classes_):
    print(f"{i} -> {label}")
```

```python
X_merged.shape
```

```python
y_clsses = label_encoder.classes_
```

```python
y_clsses
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

X_train, X_cv, y_train, y_cv = train_test_split(X_merged, y_enc, test_size=0.3)

param_grid = {'ccp_alpha': [0.1, .01, .001],
              'max_depth' : [ 15,16,17,18,19,20],
              'criterion' :['gini', 'entropy']
             }

model = DecisionTreeClassifier(random_state=1024)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose=True)
grid_search.fit(X_train, y_train)

final_model = grid_search.best_estimator_
final_model
```

```python
model = DecisionTreeClassifier(ccp_alpha=0.001, criterion='entropy', max_depth=20,
                       random_state=1024)

# Train
model.fit(X_train, y_train)

# Prediction
predictions = model.predict(X_cv)

# Evaluation
print(classification_report(y_cv, predictions))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(model, 
                           feature_names=X_merged.columns,  
                           class_names=y_clsses,  
                           filled=True, rounded=True)  

graph = graphviz.Source(dot_data, format='png')
graph.render("dt_overall_version", format='png')
graph.view()
```

```python
import os, joblib, pickle, json, datetime

# Ensure model and label encoder exist
try:
    _ = model
    _ = label_encoder
except NameError as e:
    raise RuntimeError("Model or label_encoder not found. Run training cells first.") from e

# Create folder
models_dir = 'models'
os.makedirs(models_dir, exist_ok=True)

# Filenames with timestamp for traceability
stamp = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
model_path = os.path.join(models_dir, f'decision_tree_model_{stamp}.pkl')
le_path = os.path.join(models_dir, f'label_encoder_{stamp}.pkl')
meta_path = os.path.join(models_dir, f'model_metadata_{stamp}.json')

# Save using joblib (efficient for sklearn objects)
joblib.dump(model, model_path)
joblib.dump(label_encoder, le_path)

# Minimal metadata
metadata = {
    'model_file': os.path.basename(model_path),
    'label_encoder_file': os.path.basename(le_path),
    'n_features': getattr(model, 'n_features_in_', None),
    'classes': list(label_encoder.classes_),
    'params': model.get_params(),
    'created': stamp
}
with open(meta_path, 'w') as f:
    json.dump(metadata, f, indent=2)

print('Saved:')
print('  ', model_path)
print('  ', le_path)
print('  ', meta_path)

# Quick reload smoke test
_loaded_model = joblib.load(model_path)
print('Reload check OK. Depth:', _loaded_model.get_depth())
```
